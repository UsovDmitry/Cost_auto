```python
# Дополнительно еще немного подчищу данные по стоимости авто
# уберу откровенные "выбросы": все что дешевле 120 долларов и дороже 16500
def deleted_price(column,df,k1=0.05,k2=2.5):
    qvarta = df[column].describe()
    data_25=qvarta[4]# вытаскиваем значение равное 25%
    data_75=qvarta[6]# вытаскиваем значение равное 75%
    low_25=k1*data_25 # k1 -должно быть меньше 1 т.к. задает допустимые значения до 25%, это чтобы совсем все лишнее не отсекать
    higth_75=k2* data_75 # к2 д.б. >1  т.к. задает допустимые значения свыше 75%, чтобы совсем все лишнее не отсекать
    length = 1
    index_del= []
    for index, el in enumerate(df[column]): # перебираем индексы столбца и его значения
        if low_25 > el or higth_75 < el: # все что меньше/больше межквартильго размаха - это выбросы, мусор, который уберем
            index_del.extend(df[df[column] == el].index) # формируем индексы на удаление 
    index_del= set(index_del)
    length = len(index_del)
    print('Удалено строк: {}.'.format(length))
    df.drop(index_del,axis=0,inplace=True)
```

```python
# функция заполяет пропуски значением "commit"  
def AntiMissings(df,column,commit):
    df1 = df.copy()
    df1[column] = df1[column].fillna(commit)
    df1[column] = df1[column].astype("object")
    return df1
```

```python
# сформируем функциию для расчета метрик "точности"
def rpf1_auc(target1, predict1,label1):
    print(f'Полнота {recall_score(target1, predict1)}')
    print(f'Точность {precision_score(target1, predict1)}')
    print(f'F1-мера,{label1} {f1_score(target1, predict1)}')
    print(f'AUC-ROC, {roc_auc_score(target1, predict1)}')
```

**Следующая у нас категориальная переменная содержащая существенное кол-во пропусков (~10%) это Тип кузова (VehicleType)**
В принципе ничего не машает, применить ML в заполненит пропусков и по этой переменной (утечки данных при предсказании переменной "Ремонт" не произошло, т.к. категориальные переменные не участвовали в построении модели).
Но можно попробовать пойти другим путем: каждый бренд имеет наиболее популярные модели и тип кузова которых известен. Т.е. заполним пропуски (которые сейчас у нас заменены на 'unknown') наиболее популярным типом кузова для конкретной Модели.

```python
vfg_pivot = df_new.pivot_table(index=['VehicleType'],columns=['Model'],values=['Price'], aggfunc=['count'])
vfg_pivot = vfg_pivot['count']['Price']
vfg_pivot
```

```python
def vh_fuel_gearbox(row,arg1): 
    model = row['Model']
    value = row[arg1]
    
    if value == 'unknown':
        value = vfg_pivot[model].idxmax()
    return value
```

## Обучение моделей

```python
# оформим функцию по расчету метрики RMSE
def RMSE(target,predict):
    MSE = mean_squared_error(target, predict)
    return MSE ** 0.5

RMSE_score = make_scorer(RMSE, greater_is_better=False)
```

```python
# в конечном вариант, отказался от применении метода GridSearchCV для поиска лучших параметров
# подбор займет длительное время, практически оптимальные гиперпараметры подставлю сразу "ручками"
def best_model(model, params, features_train, target_train, features_test, target_test):
    start1 = time.time()
    grid = GridSearchCV(model, param_grid = params, cv=5,
                                            scoring = RMSE_score, n_jobs=-1)
    grid.fit(features_train,target_train)
    finish1 = time.time() - start1
    start2 = time.time()
    predict = grid.predict(features_test)
    finish2 = time.time() - start2
    print (grid.best_params_)
    return finish1, finish2, RMSE(target_test, predict)
```

```python
def model_time (model,features_train, target_train,features_test, target_test,title,type_encod):
    start1 = time.time()
    model.fit(features_train, target_train)
    finish1 = time.time() - start1
    start2  = time.time()
    predict = model.predict(features_test)
    finish2 = time.time() - start2
    result = [finish1,finish2,RMSE(target_test, predict)]
    print (f'\n Реультат работы модели {title} по данным преобразованным алгоритмом - {type_encod}')
    print ('\nВремя обучения,сек.:',result[0])
    print ('Время прогнозирования,сек.:',result[1])
    print ('\nРазмер метрики RMSE:',result[2])
    return result;
```

```python
def catboost_time (type_encod,features_train, target_train,features_valid, target_valid,cat=None):
    start1 = time.time()
    model = CatBoostRegressor(loss_function='RMSE', learning_rate=0.1, 
                          random_state=12345, verbose=20,
                          iterations=200, depth=10)

    model_cb = model.fit(features_train, target_train,cat)
    finish1 = time.time() - start1
    start2  = time.time()
    predict = model_cb.predict(features_valid)
    finish2 = time.time() - start2
    result = [finish1,finish2,RMSE(target_valid, predict)]
    print (f'\nРабота модели CatBoost по данным преобразованным алгоритмом - {type_encod}')
    print ('\nВремя обучения,сек.:',result[0])
    print ('Время прогнозирования,сек.:',result[1])
    print ('\nРазмер метрики RMSE на валидационной выборке:',result[2])
    print ('\nBest score:',model_cb.best_score_)
    return result,model_cb;
```
